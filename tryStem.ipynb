```python
import pandas as pd
from stemmer import BanglaStemmer
import csv
```

```python
df = pd.read_csv("data/stim_word_frequencies.csv")
```

```python
words = df['word'].tolist()
dict = {}
```

```python
for word in words:
    temp = {}
    stm = BanglaStemmer().stem(word)
    print(stm)
    temp = {
        word : stm
    }
    dict.update(temp)
```

```python

with open('output/output.csv','w',newline='') as f:
    w = csv.writer(f)
    w.writerows(dict.items())
    #w.writeheader()
```
